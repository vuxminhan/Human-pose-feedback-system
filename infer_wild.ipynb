```python
import json
# convert to correct halpe26 input format MotionBERT
def transform_to_alphapose_format(file_path):
    # Reading the input file
    with open(file_path, 'r') as file:
        row_json = json.load(file)

    transformed_data = []
    c = 0
    for frame in row_json:
        instance = frame["instances"][0]
        transformed_instance = {}
        # Creating image_id from frame_id
        transformed_instance["image_id"] = f'{frame["frame_id"]}.jpg'
        # Setting a constant category_id
        transformed_instance["category_id"] = 1
        # Flattening keypoints and adding a default confidence score of 1.0
        keypoints = []
        points = instance["keypoints"]
        for i in range(len(points)):
            keypoints.extend([points[i][0],points[i][1], instance["keypoint_scores"][i]])  # Assuming confidence score as 1.0
        transformed_instance["keypoints"] = keypoints
        transformed_instance["score"] = instance["bbox_score"]
        transformed_instance["box"] = instance["bbox"][0]
        transformed_instance["idx"] =  [0.0]
        transformed_data.append(transformed_instance)
    print(transformed_data)
    # Returning the transformed data
    with open(file_path, 'w') as file:
      json.dump(transformed_data, file, indent=4)
json_res = vid_f.replace('mp4','json')
transform_to_alphapose_format(f'/content/Human-pose-feedback-system/data/learner/2d/{json_res}')
```
